### algorithm code


```python


import numpy as np
import time
import math


# uses euclidean distance
def calcDistance(origin, point):
    "origin is the detector and point is the space away from the detector"
    return np.sqrt(np.sum((np.array(origin[1]) - np.array(point))**2))


# knearest neighbours is needed and needs to return the set of nearest neighbours
def kNearestNeighbours(k, origin, S, dimensions):
    """
    Optimized version of kNearestNeighbours using NumPy.
    """
    distances = np.sqrt(np.sum((np.array(S) - np.array(origin[1]))**2, axis=1))
    nearest_indices = distances.argsort()[:k]
    return [(distances[i], S[i]) for i in nearest_indices]


# need to generate a new detector function
def newDetector(dimensions):
    "returns as tuple in format dimensions, age"
    return [0, np.random.random(dimensions).tolist()]

def timeDecay(time, decay, initialDecay):
    return initialDecay * math.exp(-time / decay)


# need to create the vector 1 function
def vectorOneTransform(origin, dimensions, decay, initialDecay, neighbours):
    sumOfNeighboursVector = np.sum([np.array(origin[1]) - np.array(point[1]) for point in neighbours], axis=0)
    ageDecay = initialDecay * np.exp(-origin[0] / decay)
    return origin[1] + sumOfNeighboursVector * ageDecay


# need the vector 2 function
def mewD(origin, detector, dimensions, r):
    dist = calcDistance(origin, detector[1])
    return math.exp((-(dist**2)) / 2 * r**2)


def vectorTwoTransform(origin, dimensions, decay, initialDecay, neighbours, r):
    outputVector = np.zeros(dimensions)
    ageDecay = initialDecay * np.exp(-origin[0] / decay)
    for detector in neighbours:
        dist = calcDistance(origin, detector[1])
        mewChange = np.exp(-(dist**2) / (2 * r**2))
        outputVector += (np.array(origin[1]) - np.array(detector[1])) * mewChange
    return outputVector * ageDecay



def RNS(S, r, N, max_iterations, max_age, initialDecay, decay, k):
    start_time = time.time()
    iteration = 1
    dimensions = len(S[0])
    detectors = [
        newDetector(dimensions) for i in range(N)
    ]  # randomly generated list of detectors
    while iteration <= max_iterations:
        for i, D in enumerate(detectors):
            neighbours = kNearestNeighbours(k, D, S, dimensions)
            m = np.median([a[0] for a in neighbours])

            if m < r:
                if D[0] > max_age:
                    # age is too great so replace the old detector
                    detectors[i] = newDetector(dimensions)

                else:
                    D[0] = D[0] + 1
                    D[1] = vectorOneTransform(
                        D, dimensions, decay, initialDecay, neighbours
                    )
                    detectors[i] = D
            else:
                D[0] = 0
                D[1] = vectorTwoTransform(
                    D,
                    dimensions,
                    decay,
                    initialDecay,
                    kNearestNeighbours(
                        k, D, [detector[1] for detector in detectors], dimensions
                    ),
                    r,
                )
        now = time.time()
        if now-start_time > 12:
            return [x[1] for x in detectors]
        iteration = iteration + 1
    return [x[1] for x in detectors]

```
